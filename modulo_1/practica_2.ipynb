# Práctica II de Bioinformática con BioPython
Importamos el módulo a utilizar:

```python
from Bio.Seq import Seq
```

```python
#Creamos un objeto secuencia el cual usaremos a lo largo de la práctica:
dna_seq = Seq("ATGATCTCGTAA")
```

```python
len(dna_seq)
```

```python
#Podemos obtener el complemento de la secuencia como sigue:
dna_seq.complement()
```

```python
#De igual manera el complemento reverso lo podemos encontrar de la siguiente manera:
dna_seq.reverse_complement()
```

## Transcripción y traducción.
En esta práctica analizaremos el proceso de transcripción y traducción.

```python
#Para la transcripción la realizamos de la siguiente manera:
mRNA=dna_seq.transcribe()
mRNA
```

```python
#Por otra parte la traducción se realiza de la siguiente manera:
mRNA.translate()
```

```python
#Podemos saltarnos el paso del ARN mensajero como sigue:
dna_seq.translate()
```

```python
#Podemos elegir nuestro propio codon de paro:
mRNA.translate(stop_symbol="@")
```

```python
#Podemos realizar el proceso de retro transcripción de la siguiente manera:
mRNA.back_transcribe()
```
